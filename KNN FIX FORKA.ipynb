```python
ratings_df = pd.read_csv('ratings.csv',
                         usecols=['userId', 'movieId', 'rating'],
                         dtype={'userId': 'int32', 'movieId': 'int32', 'rating': 'float32'})
```

```python
df = pd.merge(ratings_df,movies_df,on='movieId')
df #menggabungkan data yg sama yaitu Movie ID


#FAI
```

```python
combine_movie_rating = df.dropna(axis=0, subset=['title'])

movie_ratingCount = (
    combine_movie_rating
        .groupby(by=['title'])['rating']
        .count()
        .reset_index()
        .rename(columns={'rating': 'totalRatingCount'})
        [['title', 'totalRatingCount']]
)

movie_ratingCount



#dropa untuk melihat dan menghapus data yg 0
#menghitung berapa banyak rating yang diterima tiap film.
#Ngumpulin baris berdasarkan judul film, terus cuma ngambil kolom rating.
#Ngitung ada berapa rating buat setiap film.
#Biar hasilnya balik jadi tabel normal, bukan object groupby aneh.
#Kolom yang tadinya namanya rating diubah jadi totalRatingCount  jumlah rating.
#Cuma ngambil dua kolom itu, sisanya di-skip.


#11
```

```python
rating_with_totalRatingCount = combine_movie_rating.merge(movie_ratingCount, left_on = 'title', right_on = 'title', how = 'left')
rating_with_totalRatingCount.head()

#Nambahin kolom totalRatingCount ke tabel rating utama.
#Jadi tiap baris rating sekarang punya info:
#siapa user-nya
#film apa
#rating berapa
#film itu punya berapa total rating

#12
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
print(movie_ratingCount['totalRatingCount'].describe())
#Ngatur supaya angka tampil rapi (3 desimal), terus nunjukin statistik ringkas dari kolom totalRatingCount.

#mean rata rata rating
#ukuran seberapa beda-beda jumlah rating antar film.


#13
```

```python
popularity_threshold = 27  #batas minimal popularitas.
rating_popular_movie= rating_with_totalRatingCount.query('totalRatingCount >= @popularity_threshold')
rating_popular_movie.head()


#YG DIUBAHH
#14
```

```python
rating_popular_movie.shape #menghitung data rating_popular_movie

#15
```

```python
# First Lets create a Pivot matrix

movie_features_df=rating_popular_movie.pivot_table(index='title',columns='userId',values='rating').fillna(0)
movie_features_df.head()


#16

#Ubah data rating yang awalnya bentuk daftar panjang jadi matrix:
#film  user dengan nilai rating.
#Kosong diisi 0.
```

```python
from scipy.sparse import csr_matrix #csr_matrix itu cara ngubah tabel gede itu jadi versi super-hemat ruang.Daripada nyimpen ribuan angka 0, dia cuma nyimpen angka yang beneran ada.
movie_features_df_matrix = csr_matrix(movie_features_df.values) #ini eksekusi csr_matrix

from sklearn.neighbors import NearestNeighbors #Ini model yang tugasnya nyari item mirip.
model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute') #ngukur seberapa mirip arah pola rating user antar dua film. #cek semua film, satu-satu, buat nyari yang paling mirip.
model_knn.fit(movie_features_df_matrix)

#17
```

```python
movie_features_df.head() #18
```

```python
query_index = np.random.choice(movie_features_df.shape[0])
print(query_index)

distances, indices = model_knn.kneighbors(movie_features_df.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 27)

#angka di n_neighbors itu adalah k dalam knn
#YG DIUBAHHH
```

```python
movie_features_df.head()
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(movie_features_df.index[query_index]))
    else:
      print('{0}: {1}, with distance of {2}:' .format(i, movie_features_df.index[indices.flatten()[i]], distances.flatten()[i]))
```

```python
movie_name = "V for Vendetta (2006)"   # film yang mau dijadiin acuan, kode insiatif dr kelompok agar bisa milih film
query_index = movie_features_df.index.get_loc(movie_name)

distances, indices = model_knn.kneighbors(
    movie_features_df.iloc[query_index, :].values.reshape(1, -1),
    n_neighbors=27
)
```
